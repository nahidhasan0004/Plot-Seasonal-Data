```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = '/content/MONTHLY-WINTER.csv'
data = pd.read_csv(file)
```

```python
data['YEAR'] = pd.to_datetime(data['YEAR'], format='%d/%m/%Y')
```

```python
plt.figure(figsize=(12, 8))
```

```python
plt.subplot(3, 1, 1)
plt.plot(data['YEAR'], data['MAX_AVG'], color='red', label='Max Average Temperature')
plt.title('Max Average Temperature Over Time')
plt.xlabel('Year')
plt.ylabel('Max Average Temperature (°C)')
plt.grid(True)
```

```python
plt.subplot(3, 1, 2)
plt.plot(data['YEAR'], data['MIN_AVG'], color='blue', label='Min Average Temperature')
plt.title('Min Average Temperature Over Time')
plt.xlabel('Year')
plt.ylabel('Min Average Temperature (°C)')
plt.grid(True)
```

```python
plt.subplot(3, 1, 3)
plt.plot(data['YEAR'], data['RH_AVG'], color='green', label='Average Relative Humidity')
plt.title('Average Relative Humidity Over Time')
plt.xlabel('Year')
plt.ylabel('Relative Humidity (%)')
plt.grid(True)
```

```python
plt.tight_layout()
plt.show()
```
